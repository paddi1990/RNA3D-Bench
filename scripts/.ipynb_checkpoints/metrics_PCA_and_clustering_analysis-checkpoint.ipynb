```python
"""
PCA of RNA tertiary-structure evaluation metrics (metrics-as-points).

Logic (unchanged):
1) Merge dataset1 and dataset2 tables
2) Select metrics (features) and drop rows with any missing values
3) Standardize metrics across structures (z-score)
4) Perform PCA on the transposed standardized matrix (each point = one metric)
5) Plot PC1 vs PC2 with category-colored points and ellipses
"""

import os
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from plotnine import (
    ggplot, aes, geom_point, geom_text, stat_ellipse,
    theme, element_rect, element_line, element_text,
    labs, ggtitle
)

# -------------------------
# Load and merge datasets
# -------------------------
df1 = pd.read_csv("../results/metrics/dataset1_merge.csv", index_col=0)
df2 = pd.read_csv("../results/metrics/dataset2_merge.csv", index_col=0)
df = pd.concat([df1, df2], axis=0)

# -------------------------
# Define metrics and groups (keep ordering consistent)
# -------------------------
features = [
    "RMSD", "lddt", "TM-score", "GDT-TS", "LociPARSE", "RNArank",
    "INF-ALL", "f1", "precision", "recall",
    "BARNABA-eSCORE", "RNA3DCNN", "ARES", "tb_mcq", "PAMnet",
    "clash", "3drnascore", "RNA-BRiQ", "RASP-ENERGY", "DFIRE", "cgRNASP"
]
group = [
    "Geometry metrics", "Geometry metrics", "Geometry metrics", "Geometry metrics", "Deep learning metrics", "Deep learning metrics",
    "Topology metrics", "Topology metrics", "Topology metrics", "Topology metrics",
    "Geometry metrics", "Deep learning metrics", "Deep learning metrics", "Deep learning metrics", "Deep learning metrics",
    "Geometry metrics", "Energy metrics", "Energy metrics", "Energy metrics", "Energy metrics", "Energy metrics"
]

# Sanity checks (do not change computation; just fail early if inputs are inconsistent)
if len(features) != len(group):
    raise ValueError(f"`features` and `group` must have the same length. Got {len(features)} vs {len(group)}.")

missing_cols = [c for c in features if c not in df.columns]
if missing_cols:
    raise ValueError(f"Missing required metric columns in merged dataframe: {missing_cols}")

# (Note) The following line in your original code has no effect because its result is not assigned:
# df[features].dropna()

# -------------------------
# PCA (unchanged logic)
# -------------------------
black = "#222222"
gray = "#666666"

pca = PCA(n_components=2)

# Keep your exact behavior: drop any structure (row) with missing values across metrics
X = df[features].dropna()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Keep your exact behavior: PCA on metrics (transpose), so each point corresponds to one metric
pca.fit(X_scaled.T)

# Scatter dataframe: coordinates for each metric
pca_df = pd.DataFrame(pca.transform(X_scaled.T), columns=["PC1", "PC2"])
pca_df["label"] = features
pca_df["group"] = group

# -------------------------
# Plot (same content; cleaned formatting)
# -------------------------
p1 = (
    ggplot()
    + geom_point(pca_df, aes(x="PC1", y="PC2", color="group"), size=1)
    + stat_ellipse(
        pca_df, aes(x="PC1", y="PC2", color="group"),
        type="norm", level=0.6, linetype="dashed"
    )
    + geom_text(
        pca_df, aes(x="PC1", y="PC2", label="label"),
        size=6, family="Arial", color="black"
    )
    + theme(
        panel_background=element_rect(fill=gray, alpha=0),
        panel_grid_major=element_line(size=0.3, alpha=0, color=black),
        panel_grid_minor=element_line(size=0.3, alpha=0, color=black),
        panel_border=element_rect(color=black, size=1),
        axis_text=element_text(size=6, family="Arial", color="black"),
        axis_title_x=element_text(size=6, family="Arial", color="black"),
        axis_title_y=element_text(size=6, family="Arial", color="black"),
        plot_title=element_text(margin={"b": 1, "r": 0, "units": "pt"}, size=6, family="Arial", color="black"),
        figure_size=[3.1, 3.1],
        legend_title=element_text(size=6, family="Arial"),
        legend_text=element_text(size=6, family="Arial"),
        legend_background=element_rect(size=0.5, alpha=0),
        legend_position=(0.01, 0.85),
        legend_key_size=4,
    )
    + labs(x="PC1", y="PC2")
    + ggtitle("PCA plot of RNA tertiary structure evaluation metrics")
)

print(p1)

# -------------------------
# Save (mkdir if needed)
# -------------------------
out_path = "../figures/PCA_plot_of_RNA_tertiary_structure_evaluation_metrics.pdf"
os.makedirs(os.path.dirname(out_path), exist_ok=True)
p1.save(out_path)
print(f"Saved figure to: {out_path}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

from pathlib import Path
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from scipy.spatial.distance import pdist


# -----------------------------
# Publication-style settings
# -----------------------------
plt.rcParams.update({
    "font.family": "Arial",
    "font.size": 6,
    "axes.titlesize": 6,
    "axes.labelsize": 6,
    "xtick.labelsize": 6,
    "ytick.labelsize": 6,
    "legend.fontsize": 6,
    "pdf.fonttype": 42,   # embed TrueType fonts in PDF (often preferred by journals)
    "ps.fonttype": 42,
})


def plot_metric_dendrogram(
    df,
    features,
    labels=None,
    method="ward",
    metric="euclidean",
    cut_frac=0.7,
    figsize_cm=(9, 9),
    out_pdf="../figures/dendrogram_plot_of_RNA_tertiary_structure_evaluation_metrics.pdf",
    out_png=None,
    dpi=300,
    title="Hierarchical clustering of RNA structure evaluation metrics",
):
    """
    Cluster evaluation metrics (columns) using hierarchical clustering and plot a dendrogram.

    Parameters
    ----------
    df : pandas.DataFrame
        Input table containing metric columns.
    features : list[str]
        Column names (metrics) to include.
    labels : list[str] or None
        Leaf labels for the dendrogram. If None, uses `features`.
    method : str
        Linkage method (default 'ward'). Ward requires Euclidean distances.
    metric : str
        Distance metric for pdist (default 'euclidean').
    cut_frac : float
        Cut line as a fraction of the maximum linkage height (0-1).
    figsize_cm : tuple(float, float)
        Figure size in centimeters.
    out_pdf : str
        Path to save vector PDF.
    out_png : str or None
        Optional path to save PNG (useful for quick viewing).
    dpi : int
        DPI for raster outputs.
    title : str
        Plot title.
    """

    # -----------------------------
    # 1) Extract + clean
    # -----------------------------
    X = df[features].dropna(axis=0, how="any")
    if X.shape[0] < 2:
        raise ValueError(
            f"Not enough complete rows after dropna: {X.shape[0]} rows. "
            "Need >= 2 rows to compute distances."
        )

    # -----------------------------
    # 2) Standardize across samples
    #    (so metrics with large scales don't dominate)
    # -----------------------------
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X.values)  # shape: (n_samples, n_metrics)

    # -----------------------------
    # 3) Compute metricmetric distances
    #    We cluster columns (metrics), so compare columns across samples.
    # -----------------------------
    # pdist expects observations in rows, so pass transposed matrix:
    # rows = metrics, columns = samples
    dist = pdist(X_scaled.T, metric=metric)

    # Ward linkage is defined for Euclidean distance.
    if method.lower() == "ward" and metric.lower() != "euclidean":
        raise ValueError("Ward linkage requires Euclidean distances. Set metric='euclidean'.")

    Z = linkage(dist, method=method)

    # -----------------------------
    # 4) Labels handling
    # -----------------------------
    if labels is None:
        labels = list(features)
    else:
        labels = list(labels)

    if len(labels) != len(features):
        raise ValueError(
            f"labels length ({len(labels)}) must match features length ({len(features)})."
        )

    # -----------------------------
    # 5) Cut threshold + cluster ids (optional reporting)
    # -----------------------------
    cut_height = cut_frac * np.max(Z[:, 2])
    cluster_ids = fcluster(Z, t=cut_height, criterion="distance")
    # cluster_ids aligns with the *original* feature order.

    # -----------------------------
    # 6) Plot
    # -----------------------------
    fig_w_in = figsize_cm[0] / 2.54
    fig_h_in = figsize_cm[1] / 2.54

    fig, ax = plt.subplots(figsize=(fig_w_in, fig_h_in))

    dendrogram(
        Z,
        orientation="top",
        labels=labels,
        distance_sort="descending",
        show_leaf_counts=True,
        leaf_rotation=90,
        leaf_font_size=6,
        color_threshold=cut_height,
        ax=ax,
    )

    # Cut line annotation
    ax.axhline(
        y=cut_height,
        color="red",
        linestyle="--",
        linewidth=0.8,
        alpha=0.9,
        label=f"Cut line ({cut_frac:.2f}max height)",
    )

    ax.set_title(title, fontweight="bold")
    ax.set_xlabel("Evaluation metrics")
    ax.set_ylabel("Linkage distance")

    # Make the legend compact and non-intrusive
    ax.legend(loc="upper left", frameon=False)

    # Tight layout for small panels
    fig.tight_layout()

    # -----------------------------
    # 7) Save outputs
    # -----------------------------
    out_pdf = Path(out_pdf)
    out_pdf.parent.mkdir(parents=True, exist_ok=True)
    fig.savefig(out_pdf, bbox_inches="tight")

    if out_png is not None:
        out_png = Path(out_png)
        out_png.parent.mkdir(parents=True, exist_ok=True)
        fig.savefig(out_png, dpi=dpi, bbox_inches="tight")

    plt.show()

    # -----------------------------
    # 8) Return cluster membership (useful for Methods/Results)
    # -----------------------------
    # Return as {metric_label: cluster_id}
    return dict(zip(labels, cluster_ids))


# -----------------------------
# Example usage
# -----------------------------
labels = [
    "RMSD", "LDDT", "TM-score", "GDT-TS", "LociPARSE", "RNArank", "INF", "F1-score",
    "Precision", "Recall", "BARNABA-eSCORE", "RNA3DCNN", "ARES", "TB-MCQ", "PAMNet",
    "Clash", "3dRNAscore", "RNA-BRiQ", "RASP", "DFIRE", "cgRNASP",
]

cluster_map = plot_metric_dendrogram(
    df=df,
    features=features,
    labels=labels,
    cut_frac=0.7,
    figsize_cm=(9, 9),
    out_pdf="../figures/dendrogram_plot_of_RNA_tertiary_structure_evaluation_metrics.pdf",
    out_png="../figures/dendrogram_plot_of_RNA_tertiary_structure_evaluation_metrics.png",
)

# If you want to print cluster groups in the console:
from collections import defaultdict
groups = defaultdict(list)
for metric, cid in cluster_map.items():
    groups[cid].append(metric)

print("Clusters at the chosen cut line:")
for cid in sorted(groups):
    print(f"  Cluster {cid}: {', '.join(groups[cid])}")
```
